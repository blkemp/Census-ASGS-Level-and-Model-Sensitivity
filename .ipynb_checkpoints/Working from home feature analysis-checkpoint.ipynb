# Working from home: Using ML techniques to distinguish demographic features driving population behaviour
Building upon the analysis in the "ASGS Level and Modelling Sensitivity" workbook, there were some interesting features appearing in the models generated. This workbook is dedicated to further exploring this modelling in order to identify and analyse the key drivers in population behvaiours. As a starting point, the previous workbook showed that modelling at the SA2 level was best for model performance, so this is the dataset I will use for exploration. 

There were some indicative outputs from the modelling showing Age and Occupation were key features in the data, however I suspect there may be some confounding factors which are not making it clear what the real drivers are. 

Age >55 is was a feature here in terms of its dominance in effect on overall propensity to work from home. From both an SA2 and SA3 level of data, age was the dominant factor in who works from home, especially at the >50 level. That these factors are more important than the amount of factory line workers in a region or even access to the internet itself is really surprising.

Given the strength of this relationship in the above, I'm curious as to what happens when you remove age from the equation and also what happens when you add sex in as a feature. Additionally, based on the influence of "Manager" as occupation, I want to get a breakdown of the shares of this population as represented by age group and gender in order to understand if the existing "share" of managers by elder males/females may be driving the significant effects on the WFH participation rate. I.e. if it is really being driven by more older people being a manager in the first instance, or if it is being driven by other factors. Another way to support this would be to get the feature importances of solely the "Manager" columns and run simulations to see if all outputs run in the same direction.

### Metrics for evaluation
The main tools I will be using in order to explore these relationships will be "Feature Importances" of trained RandomForestRegressors used to predict Work From Home Participation Rates, as well as simulated model outcomes generated by incrementally changing the original state of these key features over a range, in order to assess both the quantum and directionality of their impacts. E.g. by changing the population of 55-65 year old managers in a given region from 0.5% to 0.7% increase or decrease the participation rate?

All inputs are standardised for the total population in a region in order to give a true participation rate which is unaffected by things like the overall employment rate and to reduce some of the noise added by the ATO for privacy purposes. That said, disentangling the unemployment rate etc. may be necessary in further analysis in order to pull out drivers.

## Define top level categories of information for training

My intuition tells me that the below features should have some form of influence on the ability of someone to work from home:   

- **Sex** - this category was not included in the original analysis, but I suspect that it may have an influence. It would be interesting to see if this influence is exhibited in preferences (working from home for convenience or driven by cultural factors) or availability (more representation in certain types of jobs with the ability to work from home).
- **Age** - I suspect there is a certain trend within particular generations to work from home, as well as a stage-of-career factor that often ties to a person's age that could be reflected. 
- **Income** - Higher skillsets have greater remuneration and flexibility to work from home.
- **Occupation** - Particular industries have the ability to work from home (software development) wheras others do not (car maintenance).
- **Family Composition** - I'm actually interested to see what this shows. Are parents more likely to work from home or people who are single?
- **Internet status** - I'd expect the ability to work from home is contingent on having internet access.   

I may choose to add in a few more tenuous links such as:
- **Building occupation status** - Are home owners or renters more likely to WFH?
- **Family with children type** - Further insight into how are parents more likely to work from home, does age of child have an influence (pre-school vs at school)?
- **Number of Bedrooms** - Are bigger houses more or less likely to entice someone to WFH?
- **Place of Birth** - Are there trends in birthplace that indicate WFH status? Cultural factors or simple immigrant vs non.
- **Country of Birth of Parents** - Potential influence on propentsity to WFH based on household cultural factors.
- **Year of Arrival** - Again, could be an interesting proxy for cultural factors influencing propensity to WFH.

## Split data by sex to determine if there are different drivers

Note this is the only feature I am able to split the "Worked from Hme" measure by, so unfortunately similar analyses for other demographic features is not possible.

```python
load_tables = ['G57','G55','G37','G33','G34','G09']
load_features = ['Age','Sex','Income','Occupation',
                 'Family Composition','Internet status',
                 'Building occupation status','Number of Bedrooms',
                 'Place of Birth']

X_sex_a2, y_sex_a2 = cnss_func.WFH_create_Xy('SA2', load_tables, load_features)
```

```python
female_cols = [fem for fem in X_sex_a2.columns if 'Female' in fem]
female_cols[:3]
```

```python
X_fem_a2 = X_sex_a2[female_cols]
X_male_a2 = X_sex_a2.drop(female_cols, axis=1)
```

```python
y_fem_a2 = y_sex_a2['Method of Travel to Work by Sex|Worked_at_home_Females_x']
y_male_a2 = y_sex_a2['Method of Travel to Work by Sex|Worked_at_home_Males_x']
```
